```python
# Iterative function to find longest decreasing subsequence of given list
def findLDS(A):
 
    # LDS[i] stores the longest decreasing subsequence of sublist
    # A[0..i] that ends with A[i]
    LDS = [[] for _ in range(len(A))]
 
    # LDS[0] denotes longest decreasing subsequence ending with A[0]
    LDS[0].append(A[0])
 
    # start from second element in the list
    for i in range(1, len(A)):
        # do for each element in sublist[0..i-1]
        for j in range(i):
            # find longest decreasing subsequence that ends with A[j]
            # where A[j] is more than the current element A[i]
            if A[j] > A[i] and len(LDS[j]) > len(LDS[i]):
                LDS[i] = LDS[j].copy()
 
        # include A[i] in LDS[i]
        LDS[i].append(A[i])
 
    # j will contain index of LDS
    j = 0
    for i in range(len(A)):
        if len(LDS[j]) < len(LDS[i]):
            j = i
 
    # print LDS
    print(*LDS[j])
 

findLDS(pi)
```
