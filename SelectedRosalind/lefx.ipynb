Enumerating k-mers Lexicographically http://rosalind.info/problems/lexf/

```python
import itertools
n = 3
lex = 'A B C D E F G'
apht = lex.split(' ')
```

```python
all_words = [''.join(i) for i in itertools.product(lex.replace(' ', ''), repeat = n)]

def is_ordered(s):
    indices = [apht.index(i) for i in s]
    #if indices == sorted(indices):
    if indices == sorted(indices) or indices == sorted(indices)[::-1]:
        return True
    else:
       return False

stt = ''
for i in all_words:
    if is_ordered(i):
        print(i, sep='\n')
        #stt += i + ' '
#print(stt)
```
