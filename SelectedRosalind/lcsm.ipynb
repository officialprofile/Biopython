Finding a shared motif

A common substring of a collection of strings is a substring of every member of the collection. We say that a common substring is a longest common substring if there does not exist a longer common substring. For example, "CG" is a common substring of "ACGTACGT" and "AACCGTATA", but it is not as long as possible; in this case, "CGTA" is a longest common substring of "ACGTACGT" and "AACCGTATA".

http://rosalind.info/problems/lcsm/

```python
with open('C:/Users/amnesia/Desktop/rosalind_lcsm.txt', 'r') as file:
    ds = file.readlines()
```

```python
sequences = []
seq_pairs = []
seq = ''
for i in ds[1:]:
    if i[0] == '>':
        sequences.append(seq)
        seq = ''
    else:
        seq += i.replace('\n', '')
sequences.append(seq)    
```

```python
def shared_sequence(s1, s2):
    temp_seq = ''
    while sequences[0][s1] == sequences[1][s2]:
        temp_seq += sequences[0][s1]
        s1 += 1
        s2 += 1
        if s1 >= len(sequences[0]) or s2 >= len(sequences[1]):
            break
    if len(temp_seq) > 5:   # arbitrary lower limit for the length of the common substring
        seq_pairs.append(temp_seq)
```

```python
for seq1 in range(len(sequences[0])):
    for seq2 in range(len(sequences[1])): 
        if sequences[0][seq1] == sequences[1][seq2]:
            shared_sequence(seq1, seq2)

seq_pairs = list(set(seq_pairs))
seq_pairs.sort(key = len)
common = dict(zip(seq_pairs, [0]*len(seq_pairs)))

for i in sequences:
    for j in seq_pairs:
        if i.find(j) > -1:
            common[j] += 1
```
